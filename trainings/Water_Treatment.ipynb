# Water Treatment

Treated water can be reused for agricultural purposes.  WAS allows the user to model construction of new water treatment facilities that increase the supply of recycled water.  Recycled water can only be used for agricultural demand, but in some cases may be  used as a substitute for fresh water.

## Adding a new Plant

Adding a new treatment plant increases the supply of recycled water available by adding a new step to the supply function.

```python
import econgraphs
from ipywidgets import interact
from econgraphs import graphMaker, highAltMarket
```

```python
@interact(cost_mcm = (.5,6,.1),capacity = (.2,2,.1))
def drawNewLinkage(cost_mcm, capacity):
    graphMaker.drawSupplyShift(newPoint = (cost_mcm,capacity))
```
