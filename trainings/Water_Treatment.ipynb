# Water Treatment

All Treated water can be reused for agricultural purposes.  WAS allows the user to model construction of new water treatment facilities that increase the supply of recycled water.  Recycled water can only be used for agricultural demand, but in some cases may be  used as a substitute for fresh water.

## Baisc Water Treatment

Water that is not recycled must still be treated before being discharged. Treatment of water after use has no direct impact on demand, but untreated water is a serious threat to the environment.  For this reason, there is a default cost of water treatment in WAS, which can be adjusted using the subsidy/tax social policy tool.  To open this tool, use the "edit" button next to Environmental Charges and Setasides:

![env window][env-window]

and select a country.  You will see the default tax on water usage in the urban % tax and industry % tax columns, which are set to represent the cost of water treatment in each district. 

![treatment][treatment]

The user can modify the cost of water treatment for each district, or apply the same cost to all districts in a country using the right hand set all values tool on the right side

## New Recycling Plant

Adding a new treatment plant increases the supply of recycled water available to agriculture by adding a new step to the supply function.

```python
%%capture
import econgraphs
from ipywidgets import interact
from econgraphs import graphMaker, highAltMarket
```

```python
@interact(cost_mcm = (.5,6,.1),capacity = (.2,2,.1))
def drawNewLinkage(cost_mcm, capacity):
    graphMaker.drawSupplyShift(newPoint = (cost_mcm,capacity))
```

Even though agricultural buyers are the only buyers who can use recycled water, urban and industrial buyers benefit too: When agricultural buyers switch to using recycled water, more fresh water becomes available for other buyers.

#### Cost

The cost is **not** the construction cost.  Construction costs needs to be taken into account separately.  

The cost below represents the cost per m<sup>3</sup> of treating water to make it usable for agriculture.  The cost of treating water so that is safe for disposal should not be included in this number.

![rec cost][rec-cost]

[rec-cost]:https://github.com/akaufman10/was/blob/master/static/screenshots/rec-cost.png?raw=true

#### Maximum Percentage

The maximum percentage is determine the amount of water that can be recycled from fresh water waste, based on the usage in a district.  If a district uses 40 milion m<sup>3</sup> of water and the plant has an 70% maximum percnetage, it will produce 28 million m<sup>3</sup> of recycled water.

![rec pct][rec-pct]

[rec-pct]:https://github.com/akaufman10/was/blob/master/static/screenshots/rec-pct.png?raw=true

#### Maximum Capacity

The capacity is how much water the plant can treat.  If the plant described above has a capacity of only 25 m<sup>3</sup>, it will produce only 17.5 million m<sup>3</sup>, as it is further limited by the maxiumum percentage of water it can recycle.
