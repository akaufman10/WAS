# Environment and Sustainability

There are several ways to include environmental and sustainability policy into WAS. The three methods are basically 
* limiting water usage to sustainabile levels.
* allocating water to be "set aside," or saved for environmental purposes.
* imposing additional taxes on water to compensate for environmental damage.

These methods may be used separately or together


## Limiting Water Usage

The most basic way to incorporate sustainabile water policy in WAS is to limit the amount of water that may be extracted from a given source.  Outside data is needed to determine the level of sustainabile usage.  

### Theory

Limiting water use from a source to a sustainable level is just like lowring the capacity of a water source.  Reducing the water available from a source shrinks the cooresponding step in the supply function, as shown below.

```python
%%capture
import econgraphs
from ipywidgets import interact
from econgraphs import graphMaker, highAltMarket
```

```python
@interact(step = (1,3,1),reduction = (.2,2.9,.1))
def drawNewLinkage(step,reduction):
    reduction = -reduction
    graphMaker.drawSupplyShift(newPoint = (step*3,reduction))
```

Reducing the supply of water available will always decrease social welfare in WAS as society will have less water to use.

## Setasides (Reseved Water)

Sometimes water is needed for an environmental purpose, such as keeping a river flowing or providing a natural sanctuary.  WAS allows the user to set aside a quantity of water for such a purpose - this water will not be consumer by the public and will not contribute to the social welfare that WAS calculates.  

### Theory

A setaside does not shift the supply or demand curved - it just removes the net benefit that would normally go to the users for a specific quantity of water, as shown below

```python
graphMaker.annotate()
@interact(setaside=(0,2,.1),waterAvailable = (2,4,.1),annotate=(True))
def drawNewLinkage(setaside,waterAvailable,annotate):
    newset = graphMaker.market.equilibriumQ - (waterAvailable - setaside)
    if newset > 0:
        graphMaker.drawSetAside(setaside = newset, annotate=annotate)
```

Note that a setaside reduces surplus **only** when the supply of water is limited. If there is enough water, the setaside does not change the equilibrium level of water supplied.
